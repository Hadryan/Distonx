```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = '~/data/distonx.csv'
data = pd.read_csv(PATH)
```

```python
data
```

```python
data['target'] = (data['depth_bid_price_1'] + data['depth_ask_price_1']) / 2
```

```python
data.loc[range(0, 200, 11)]
```

```python
data['target'].shift(-110).loc[range(0, 100, 11)]
```

```python
data['shift_10_s'] = data['target'].shift(-110)
```

```python
data.dropna(inplace=True)
```

```python
btcusd = data[data['currency_pair'] == 'btcusdt']
```

```python
btcusd['index'] = btcusd.index
```

```python
melt = pd.melt(btcusd, id_vars=['index'], value_vars=['target', 'shift_10_s'])
```

```python
melt
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.lineplot(data=melt[melt['index'] < 1000], x='index', y='value', hue='variable', ax=ax)
```

```python
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
model.fit(btcusd.drop(['target', 'shift_10_s', 'currency_pair'], axis=1), btcusd['shift_10_s'])
```

```python
btcusd['prediction'] = model.predict(btcusd.drop(['target', 'shift_10_s', 'currency_pair'], axis=1))
```

```python
melt = pd.melt(btcusd, id_vars=['index'], value_vars=['target', 'shift_10_s', 'prediction'])
fig, ax = plt.subplots(figsize=(15, 10))
sns.lineplot(data=melt[melt['index'] < 1000], x='index', y='value', hue='variable', ax=ax)
```
